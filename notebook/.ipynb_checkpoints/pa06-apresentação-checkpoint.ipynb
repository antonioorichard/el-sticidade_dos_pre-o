REFLEXÃO p/ trabalhar com os NANs

**Objetivo do trabalho:** Calcular a elasticidade dos preços ! 

Formas de tratar os NAN:

    * Deletar ( df1.dropna() )
        * + (vantagem        -  Simples
        * - (Desvantagem )   -  Perda de informação
        * Goal               -  Depende da quantidade 
   
    * Preencher
        * + (vantagem        -  Mantém o tamanho original dos dados
        * - (Desvantagem )   -  Caso a distribuição n seja normal, pode fica com muito viés!
        * Goal               -  !
    
    * Interpolar
        * + (vantagem        -  Mantém a integridade de seres temporais 
        * - (Desvantagem )   -  Pode distorcer com informações falsas
        * Goal               -  
    
    * Modelos Estatísticos
        * + (vantagem        -  Abordagem mais sofisticada 
        * - (Desvantagem )   -  Reque muito recurso do computador e pode ser impreciso
        * Goal               -  

Conclusão: Melhor caminho baseado que nas colunas que tem não são tão relevante para o objetivos deletar elas é a melhor opção

```python
df_speaker.isnull().sum()
```

```python
dataset = ( df_speaker.groupby(['name', 'week_number'  ]).
           agg({'disc_price':'mean', 'date_imp':'count'}).
           reset_index() )
dataset.head(15)
# date_imp é quantidade de produto vendido na aquela semana
```

**Bem o processo aqui tem defeito de esquecer a quantidade vendida ou seja a demanda quando é feito pivot ou rotação das tabela!**

```python
# criando dataframe dos valores
x_price = dataset.pivot( index = 'week_number', columns = 'name', values = 'disc_price')

x_price = pd.DataFrame( x_price.to_records())
x_price.head(20)
# Problema deste processo é perca da quantidade vendida na semana!
```

**Agora vem a tentativa de resolução para não esquecer a quantidade vendida em cada dia da semana, criando y_demand**

```python
y_demand = dataset.pivot(index = 'week_number', columns = 'name', values = 'date_imp')
y_demand = pd.DataFrame( y_demand.to_records())
y_demand.head()
```

## 4.2 MELHORIA NO x_price e y_demand Pana NaN
* **Como não tem informação do preço na semana que não vende podemos presumir que o preço é o mesmo da semana anterior!**
* **Olha, no y_demand, onde contém NaN, subtitui por zero, pois foi vendido zero quantidade em tal semana**

## 5.4 Replace NAN

```python
x_price.head(15)
```

```python
# Experimento controlado com ajuste para valores posição 1 NaN
# Ideia

columns_list = list(x_price.iloc[:,:].columns)
week_list    = list(x_price.iloc[:,0])
x_price_ = x_price.copy()
for coluna in columns_list:
    position = 0
    print (f'coluna:',  {coluna} )
    for   week in week_list:
               
        test   = x_price_.loc[x_price_['week_number'] == week, coluna ].isnull().bool() ; 
        print(f'week: {week}, condição: {test}, Posição: {position}' )
        if test == True:
            print('Entrou' )
            if position != 0:
                valor =  x_price_.loc[x_price_['week_number'] == week_list[position-1], coluna] 
                x_price_.loc[x_price_['week_number'] == week, coluna ] = valor.values[0]
            else:
                print('linha inicial com NaN')
                somador   = 1 
                position1 = position
                test1     = test
                
                while test1 == True :
                    print(f'Dentro do while, position1:{position1}, test1: {test1}')
                    position1 = position1 + 1
                    test1    = x_price_.loc[x_price_['week_number'] == week_list[position1], coluna ].isnull().bool()
                     
                    
                valor = x_price_.loc[x_price_['week_number'] == week_list[position1], coluna] 
                x_price_.loc[x_price_['week_number'] == week, coluna ] = valor.values[0]
                print(f'valor: {valor}' )
                                  
# Tem comparar com o trabalho versão anterior ver o se esse tipo de ajuste trouce alguma vantagem em R² p-value
# Colocar na  apresentação também! ou Readme!
        position = position + 1
x_price = x_price_        
x_price.head()
```

```python
results_values_speaker = {
    "name": [], 
    "price_elasticity": [], 
    "mean_price": [], 
    "mean_quantity": [],  # Demanda
    "intercept": [],      # coeficiente linear
    "slope": [],          # coeficicente angular
    "rsquared": [],       # r²
    "p_value": [],        # significância
}

for column in x_price.columns[1:]:
    column_points = []
    for i in range( len(x_price[column])):
        column_points.append((x_price[column][i], y_demand[column][i]))
    df = pd.DataFrame( list(column_points), columns = ["x_price", "y_demand"])

    x_speaker = df["x_price"]
    y_speaker = df["y_demand"]
    X_speaker = sm.add_constant(x_speaker)
    model     = sm.OLS(y_speaker, X_speaker)
    results   = model.fit()

    if results.f_pvalue < 0.15:
        rsquared         = results.rsquared
        p_value          = results.f_pvalue
        intercept, slope = results.params
# Nos cálculos a seguir de mean_price, e mean_quantity estarão poluídos
        # com valores tando na média de preço quanto na quantidade sendo que não vendeu !
        # O que irá distorcer ainda mais, não só devido aos NaN's e minha escolha de como corrigir!.
        mean_price       = np.mean(x_speaker)
        mean_quantity    = np.mean(y_speaker) 

        price_elasticity = slope * (mean_price / mean_quantity)

        results_values_speaker['name']            .append(column)
        results_values_speaker['price_elasticity'].append(abs(price_elasticity))
        results_values_speaker['mean_price']      .append(mean_price)
        results_values_speaker['mean_quantity']   .append(mean_quantity)
        results_values_speaker['intercept']       .append(intercept)
        results_values_speaker['slope']           .append(slope)
        results_values_speaker['rsquared']        .append(rsquared)
        results_values_speaker['p_value']         .append(p_value)
       

df_elasticity = pd.DataFrame(results_values_speaker)
df_elasticity
```

```python
resultado_experimento = {
    "name"                :[],
    "price_elasticity"    :[],
    "operacao"            :[],
    "antigo_total"        :[], 
    "novo_total"          :[],
    "delt_gain"           :[],
    "delt_gain_%"         :[],
    "delt_demanda"        :[], 
    "total"               :[]
}

for i in range( len( df_elasticity)):
    name           = df_elasticity["name"][i]
    preco_antigo   = round(df_elasticity["mean_price"][i],2)
    tipo           = "desconto"
    variacao_price = 15 
    novo_preco     = preco_antigo * 0.85
    antiga_demanda = df_elasticity["mean_quantity"][i]
    e              = round(df_elasticity["price_elasticity"][i],2) 
    
    if e > 1: # Em casos que clientes são sensíveis a variação de preço
       if tipo ==  "desconto":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * ( (variacao_demanda / 100) + 1.2)
       if tipo == "aumento":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * (1 - (variacao_demanda / 100) )

    elif (e == 1):
    # se aumenta um preço a queda tende a ser proporcional ao aumento
        if tipo ==  "desconto":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * ( (variacao_demanda / 100) + 1)
        if tipo == "aumento":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * (1 - (variacao_demanda / 100) )
            
    elif (e < 1):
       # caso da inelástico a variação para aumento é pouco
        if tipo ==  "desconto":
            variacao_demanda = e * variacao_price/ 2.5
            nova_demanda     = antiga_demanda * ( (variacao_demanda / 100) + 1)
        if tipo == "aumento":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * (1.2 - (variacao_demanda / 100) )

    
        
        
# Faturamento
    novo_total     = round(novo_preco  * nova_demanda,  2) # Novo Faturamento
    antigo_total   = round(preco_antigo * antiga_demanda,2) # Antigo Faturemento
# /Faturamento

# Variação percentual
    delt_gain           = round(novo_total - antigo_total, 2) # variação do ganho

    variacao_percentual = round( ((delt_gain ) / antigo_total ) * 100, 2) # ganho porcentagem

    delta_demanda       = round( nova_demanda - antiga_demanda , 2) 
    
# /Variação percentual

    resultado_experimento["name"].                append(df_elasticity["name"][i])
    resultado_experimento["price_elasticity"].    append(e)
    resultado_experimento["operacao"].            append( tipo)
    resultado_experimento["antigo_total"].        append(antigo_total)
    resultado_experimento["novo_total"].          append(novo_total)
    resultado_experimento["delt_gain"].           append(delt_gain)
    resultado_experimento["delt_gain_%"].         append(variacao_percentual)
    resultado_experimento["total"].               append('total')
    resultado_experimento["delt_demanda"].        append(delta_demanda) # variação na demanda

resultado_experimento_random = pd.DataFrame( resultado_experimento)
resultado_experimento_random
```

```python
# Trabalho inteligente com os preços!
resultado_experimento = {
    "name"                :[],
    "price_elasticity"    :[],
    "operacao"            :[],
    "antigo_total"        :[], 
    "novo_total"          :[],
    "delt_gain"           :[],
    "delt_gain_%"         :[],
    "delt_demanda"        :[], 
    "total"               :[]
}

for i in range( len( df_elasticity)):
    name           = df_elasticity["name"][i]
    preco_antigo   = round(df_elasticity["mean_price"][i],2)
    tipo           = "aumento"
    variacao_price = 15 
    novo_preco     = preco_antigo * 0.85
    antiga_demanda = df_elasticity["mean_quantity"][i]
    e              = round(df_elasticity["price_elasticity"][i],2) 
    
    if e > 1: # Em casos que clientes são sensíveis a variação de preço
       tipo  = "desconto"
       if tipo ==  "desconto":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * ( (variacao_demanda / 100) + 1.2)
       if tipo == "aumento":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * (1 - (variacao_demanda / 100) )

    elif (e == 1):
    # se aumenta um preço a queda tende a ser proporcional ao aumento
        if tipo ==  "desconto":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * ( (variacao_demanda / 100) + 1)
        if tipo == "aumento":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * (1 - (variacao_demanda / 100) )
            
    elif (e < 1):
       # caso da inelástico a variação para aumento é pouco
        tipo = "aumento"
        if tipo ==  "desconto":
            variacao_demanda = e * variacao_price/ 2.5
            nova_demanda     = antiga_demanda * ( (variacao_demanda / 100) + 1)
        if tipo == "aumento":
            variacao_demanda = e * variacao_price
            nova_demanda     = antiga_demanda * (1.2 - (variacao_demanda / 100) )

    
        
        
# Faturamento
    novo_total     = round(novo_preco  * nova_demanda,  2) # Novo Faturamento
    antigo_total   = round(preco_antigo * antiga_demanda,2) # Antigo Faturemento
# /Faturamento

# Variação percentual
    delt_gain           = round(novo_total - antigo_total, 2) # variação do ganho

    variacao_percentual = round( ((delt_gain ) / antigo_total ) * 100, 2) # ganho porcentagem

    delta_demanda       = round( nova_demanda - antiga_demanda , 2) 
    
# /Variação percentual

    resultado_experimento["name"].                append(df_elasticity["name"][i])
    resultado_experimento["price_elasticity"].    append(e)
    resultado_experimento["operacao"].            append( tipo)
    resultado_experimento["antigo_total"].        append(antigo_total)
    resultado_experimento["novo_total"].          append(novo_total)
    resultado_experimento["delt_gain"].           append(delt_gain)
    resultado_experimento["delt_gain_%"].         append(variacao_percentual)
    resultado_experimento["total"].               append('total')
    resultado_experimento["delt_demanda"].        append(delta_demanda) # variação na demanda

resultado_experimento_c = pd.DataFrame( resultado_experimento)
resultado_experimento_c
```

```python
def crossprice(df_x, df_y, column_name):
    # Pegando todos os valores de x_price
    new_df = df_x.copy()

    # Pegando os valores y_demand com nome da coluna
    new_df['y_value-'+ column_name] = df_y[column_name]

    multi_xvalues = new_df.loc[:, new_df.columns[1:-1]]
    multi_yvalues = new_df.loc[:, new_df.columns[-1]]

    # Obter o valor medio do preço do produto
    mean_values   = np.mean(multi_xvalues)
    
    # Obter o valor medio demanda 
    mean_yvalues  = np.mean(multi_yvalues)

    # Regressão Linear

    x     = sm.add_constant(multi_xvalues)
    model = sm.OLS(multi_yvalues, x, missing = 'drop')
    result= model.fit()

    # Obtendo os resultados 
    results_summary = result.summary()

    # p-valores para coeficiente
    p_value         = result.pvalues

    # Transformando os resultados em uma dataframe

    results_as_html = results_summary.tables[1].as_html()
    new_dataframe   = pd.read_html( results_as_html, header = 0, index_col = 0)[0]

    # Adicionando o p-valor ao dataframe
    new_dataframe['p_value'] = p_value

    # Definindo o nome do produto como index
    new_dataframe.index.name = 'name'
    new_dataframe.reset_index()

    # Calculando a elasticidade cruzada
    new_dataframe['mean']             = mean_xvalues
    new_dataframe['price_elasticity'] = round(( new_dataframe.coef) * (new_dataframe['mean']/mean_yvalues), 2)

    new_dataframe                     = new_dataframe.reset_index()
    pvalue_signicant                  = new_dataframe['p_value']

    # Verificando a hipótese nula (inclinação por produto)
    new_dataframe[column_name+'CPE'] = np.where(( pvalue_signicant > 0.05), 'No Effect', new_dataframe['price_elasticity'])
    new_dataframe.dropna()
    return new_dataframe[[ 'name', column_name + 'CPE' ]]
```

```python
result_df = pd.DataFrame()
for column in x_price.columns[1:]:
    result_df[['name', column + 'CPE']] = crossprice(x_price, y_demand, column)
```
